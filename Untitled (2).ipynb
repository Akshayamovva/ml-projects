```python
columns = ['value', 'classification','Execution Price', 'Size Tokens', 'Size USD', 'Side', 'Direction', 'Fee', 'Closed PnL', 'Crossed']
```

```python
df = df_merged[columns]
```

```python
df
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
y = df['classification']
```

```python
x = df.drop(['classification'], axis =1 )
```

```python
x.dtypes
```

```python
x['Direction'] = le.fit_transform(x['Direction'])
x['Side'] = le.fit_transform(x['Side'])
```

```python
x['Crossed'] = le.fit_transform(x['Crossed'])
```

```python
y
```

```python
x
```

```python
y = le.fit_transform(y)
```

```python
from sklearn.preprocessing import MinMaxScaler
mn = MinMaxScaler()
x['value'] = mn.fit_transform(x[['value']])
x['Execution Price'] = mn.fit_transform(x[['Execution Price']])
x['Size Tokens'] = mn.fit_transform(x[['Size Tokens']])
x['Size USD'] = mn.fit_transform(x[['Size USD']])
```

```python
x.head()
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(x,y, test_size=0.3)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model = RandomForestClassifier()
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import classification_report
```

```python
y_test
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test, y_pred)
```
